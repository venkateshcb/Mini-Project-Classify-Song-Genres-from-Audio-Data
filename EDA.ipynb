```python
import numpy as np
import os
import tensorflow as tf
import matplotlib.pyplot as plt
from glob import glob
import librosa
import pandas as pd
from librosa import load, display
```

```python
classes = os.listdir("data/genres_original")
classes
```

```python
hiphop = glob(f"data/genres_original/hiphop/*.*")
rock = glob(f"data/genres_original/rock/*.*")
```

#   FOR HIPHOP

```python
wav1, sample_rate = librosa.load(hiphop[0])
plt.figure(figsize=(15,5))
librosa.display.waveshow(wav1, sr = sample_rate)
```

MEL FREQUENCY CEPTRAL CO-EFFICIENTS FOR HIPHOP

```python
mfcc = librosa.feature.mfcc(wav1, sr = sample_rate)
mfcc.shape
plt.figure(figsize=(15,5))
librosa.display.waveshow(mfcc, sr = sample_rate)
```

SPECTROGRAM

```python
plt.figure(figsize=(15,5))
D = librosa.amplitude_to_db(np.abs(librosa.stft(wav1)), ref=np.max)
img = librosa.display.specshow(D, y_axis='linear', x_axis='time',
                               sr=sample_rate)
```
